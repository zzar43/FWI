```julia
# Test
pml_len = 10;
pml_alpha = 200;

Nx = 50;
Ny = 101;
h = 0.1;

Nt = 1000;
dt = 0.001;

c = 10 * ones(Nx,Ny);
c[1:20,1:50] = 40;
c[1:20,50:end] = 40;
# c[25:26,50:end] = 40;
# c[30:33,:] = 5;

print(maximum(c)*dt/h)

source_coor = [49 51];
receiver_coor = 50 * ones(Int,50,2);
receiver_coor[:,2] = 1:2:99;

source_func = zeros(1,Nt);
source_func[1,:] = source_ricker(Nt, dt, 100, 0.02);
# source_func[2,:] = source_ricker(Nt, dt, 100, 0.02);
# source_func[1,:] = source_sin(Nt, 10);
# source_func[2,:] = source_sin(Nt, 10);
# source_func[1,:] = source_spike(Nt, 20);
# source_func[2,:] = source_spike(Nt, 20);

t = linspace(0,(Nt-1)*dt,Nt);
p1 = plot(t,source_func[1,:]);

p2 = heatmap(c);
scatter!(receiver_coor[:,2], receiver_coor[:,1])
scatter!(source_coor[:,2], source_coor[:,1])

plot(p1,p2)
```

```julia
uu, snaps_uu, rec_data = wave_solver_2d_pml(c,Nx,Ny,h,Nt,dt,pml_len,pml_alpha,source_coor,source_func,receiver_coor);
```
