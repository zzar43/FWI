# 1 Dimensional Helmholtz Equation

```julia
using Plots;
pyplot()
```

```julia
N = 100;
m = ones(N);
m[50:end] = 2;
omega = 2;
dx = 1/N;
x = linspace(0,1-dx,N);
f = zeros(N);
f[30] = 1;

d = -2*ones(N-2);
dl = ones(N-3);
du = ones(N-3);

u = zeros(N);
```

```julia
A = 1/(2*dx)*Tridiagonal(dl,d,du)
B = omega^2.*Diagonal(m[2:end-1])
A = A + B
heatmap(A)
```

```julia
cond(Array(A),2)
# size(A)
```

```julia
u[2:end-1] = -1*A\(f[2:end-1])
```

```julia
plot(x,u)
```
