# Sensitivity kernal shown by single source and receiver
Sensitivity kernal with sin function source. In a homogeneous background.

```julia
Nx, Ny = 101, 101; # grid number
# True velocity model
vel_true = 2000*ones(Nx,Ny);
vel_true[:,75:end] = 2500;
vel_init = 2000*ones(Nx,Ny);

# Time
sample_fre = 1000; # Hertz
dt = 1/sample_fre;
Nt = 2000;
t = linspace(0,(Nt-1)*dt,Nt);

source_num = 1;
source_coor = zeros(Int,source_num,2);
for i = 1:source_num
    source_coor[i,1] = 20;
    source_coor[i,2] = 30;
end
# source_vec0 = 100*source_ricker(25, 0.05, t);
source_vec0 = 100*source_sine(50, t);
source_vec = zeros(source_num, Nt);
for i = 1:source_num
    source_vec[i,:] = source_vec0;
end
# Receiver
receiver_num = 1;
receiver_coor = zeros(Int,receiver_num,2);
for i = 1:receiver_num
    receiver_coor[i,1] = 80;
    receiver_coor[i,2] = 30;
end
include("time_test1.jl")
```
